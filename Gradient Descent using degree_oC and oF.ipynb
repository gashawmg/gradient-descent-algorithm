```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generate some temperature data (in Celsius and Fahrenheit scales )

```python
deg_celsius = [] 
deg_fahreheit = []
for i in range(50):
    deg_celsius.append(i)
    deg_fahreheit.append(1.8 * i + 32)     
```

```python
# oF = w*oC + b
# Numpy arrays are memory efficient and faster for computation than lists
C = np.array(deg_celsius,dtype=np.float64)  # temperature in degree celsius
F = np.array(deg_fahreheit,dtype=np.float64) # temperature in degree Fahrenheit
```

```python
# Plot the graphs
plt.figure(figsize=(6,4))
plt.plot(C,F,'ro')
plt.ylabel('Fahrenheit Scale ')
plt.xlabel('Celsius Scale')
plt.title('Fahrenheit and Celsius Relationship')
plt.ylim(0,200)
```

## Loss functions and gradient descent.

<div style="background-color:#daf0ef">
Mean Squared Error, MSE, is used as a loss function or objective functions in supervised learning - regression problems<br>
<b>MSE o</b> is defined as:
$$\ L=\frac{1}n \sum_{i=1}^n (y_i -\hat{y_i})^2 $$
Where:<br/>
y<sub>i</sub> is the i<sup>th</sup> observed value.<br/>
$\hat{y_i}$ is the corresponding predicted value.<br/>
n = the number of observations.<br/>
For <i><b>linear regression</b></i> with one variable: $\hat{y_i} = wx_i + b$
$$\\L=\frac{1}n \sum_{i=1}^n (y_i -(wx_i + b))^2$$ 
The partial derivative of <b>L</b> with respect to <b>w</b>:
$$ \frac{\partial L}{\partial w} = \frac{1}n \sum_{i=1}^n-2x_i (y_i -(wx_i + b))$$
The partial derivative of <b>L</b> with respect to <b>b</b>:
$$ \frac{\partial L}{\partial b} = \frac{1}n \sum_{i=1}^n-2 (y_i -(wx_i + b))$$

$\hat{y_i} = wx_i + b$  =>   $\hat{F_i} = wC_i + b$


<br/><br/>
Reference:[Hundred-Page-Machine-Learning](https://www.amazon.com/Hundred-Page-Machine-Learning-Book/dp/199957950X)

</div>

```python
n = len(C) # 50 data points
oF_predictions = []
learning_rate = 0.01

w = 0.0 # weight/slope 
b = 0.0 # bias/y_intercept

for epoch in range(120):
    dw = 0.0
    db = 0.0

    for i in range(n): # 50 iterations(n = 50), inner loop
        # gradients are accumulated
        dw = dw + (-2 * C[i] * (F[i] - (C[i] * w + b)))/n # C[0] = 0
        db = db + (-2 * (F[i] - (C[i] * w + b)))/n # F[0]=32 

        # update variables
        w = w - dw * learning_rate
        b = b - db * learning_rate

        pred = C[i] * w + b
        oF_predictions.append(pred)
    print(f'epoch:{epoch} dw:{dw:.3f}   db: {db:.3f} w: {w:.3f} b: {b:.3f} pred: {pred:.3f}') 
```

```python
plt.plot(C,F,'yo')
plt.plot(C,oF_predictions[:50],'ro')
plt.xlabel('oC')
plt.ylabel('oF')
plt.legend(['Ground-truth values','First epoch Prediction'])
```

```python
def gradient_descent(C, F, w, b, learning_rate, epochs):
    w_gradients=[]
    b_gradients= []
    w_values = []
    b_values = []
    MSE = []
    oF_predictions=[]
    n = len(C)
    epoch_ = []
    for epoch in range(epochs):
        #----- in each epoch dw and db gradients should be intialized
        #----- for  w and b values in the previous epoch will be used as initial values
        dw = 0.0
        db = 0.0
        for i in range(n):
            dw = dw + (-2 * C[i] * (F[i] - (C[i] * w + b)))/n
            db = db + (-2 * (F[i] - (C[i] * w + b)))/n # F[0]=32 
            loss = ((F[i]-(C[i] * w + b))**2).sum()/n
            pred = C[i] * w + b
            oF_predictions.append(pred)
            
            # update w and b
            w = w - dw * learning_rate
            b = b - db * learning_rate
            
        #------ Lists append the last iteration of each epoch----   
        
        b_gradients.append(db)
        w_gradients.append(dw)
        w_values.append(w)
        b_values.append(b)
        MSE.append(loss)
        epoch_.append(epoch) 
        print(f'epoch:{epoch} dw:{dw:.2f} db: {db:.2f} w: {w:.5f} and b: {b:.5f} MSE:{loss:.12f}')
    return b_gradients,w_values,b_values,w_gradients,MSE,oF_predictions,epoch_  
```

```python
b_gradients,w_values,b_values,w_gradients,MSE,oF_predictions,epoch_ = gradient_descent(C, F, 0, 0, 0.01, 120)
```

```python
# Plot the graphs
plt.figure(figsize=(8,5))
plt.plot(C, oF_predictions[:50],'y') # first epoch predictions
plt.plot(C, oF_predictions[50:100],'g')# second epoch predictions
plt.plot(C, oF_predictions[100:150],'m')# third epoch predictions
plt.plot(C, oF_predictions[150:200],'c')# fourth epoch predictions
plt.plot(C,oF_predictions[200:250],'b') # fifth epoch predictions
plt.plot(C,oF_predictions[-50:],'ro') # last epoch predictions
# plt.plot(C, F,'r',)
plt.ylabel('Fahrenheit Scale',color='red')
plt.xlabel('Celsius Scale', color='red')
plt.title('Degree Fahrenheit Predictions')
plt.legend(['First epoch Prediction','Second epoch Prediction', 
            'Third epoch Prediction','Fourth epoch Prediction',
            'Fifth epoch predictions','last epoch prediction'])
```

```python
# Plot MSE and w-values for every epoch
plt.plot(w_values, MSE,'ob--')
plt.ylabel('Mean Squared Error',color='red')
plt.title('MSE vs Slopes(w_values)')
plt.xlabel('w',color='red')
```

```python
# Plot MSE and b_values for every epoch

plt.plot(b_values, MSE,'ob--')
plt.ylabel('Mean Squared Error(MSE)',color='red')
plt.xlabel('b',color='red')
```

```python
# gradients with respect to w
plt.plot(w_gradients, MSE,'ob--')
plt.ylabel('Mean Squared Error(MSE)',color='red')
plt.xlabel('w-gradients',color='red')
```

```python
# gradients with respect to w
plt.plot(b_gradients, MSE,'ob--')
plt.ylabel('Mean Squared Error', color='red')
plt.xlabel('b_gradients',color='red')
```

## Effects of learning rate

i) Bigger learning rate: if you pass 0.1 learning rate, error explodes and you will encouter overflow 

```python
b_gradients,w_values,b_values,w_gradients,MSE,oF_predictions,epoch_ = gradient_descent(C, F, 0, 0, 0.1, 120)
```

```python
# Plot the relationship between epochs or number of iterations and 
plt.plot(epoch_, MSE,'ob--')
plt.ylabel('Mean Squared Error')
plt.xlabel('Epochs')
```

ii) Smaller learning rate: if you pass 0.001 learning rate(1/10 of the reasonable(0.01)  learning rate), it requires a lot of training to achieve the same result 

```python
b_gradients,w_values,b_values,w_gradients,MSE,oF_predictions,epoch_ = gradient_descent(C, F, 0, 0, 0.001, 1200)
```
